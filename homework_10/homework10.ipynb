# 10.2.1.На какой дистанции Jesse Owens выйграл медаль?

```python
distances = ('100m', '200m', '1500m', '800m', '4x100m relay', '4x200m relay')
jesseevent = [d for d in oo[(oo['Athlete'] == 'OWENS, Jesse')]['Event'] if d in distances]

for i in jesseevent:
    print(f"Jesse Owens выйграл медаль на дистанции {i}")
```

# Какая страна выйграла большинство золотых медалей мужчинами в бадминтоне?

```python
badmenton = oo[(oo['Medal'] == 'Gold') & (oo['Gender'] == 'Men') & (oo['Sport'] == 'Badminton')]
top_badmentons = badmenton.NOC.value_counts()
print(f"Страна, выигравшая боль-во золотых медалей мужчинами в баментоне - {top_badmentons.index[0]}; кол-во медалей :{top_badmentons.values[0]} ")
```

# Отсортировать результат(топ_1 страну) в алфавитном порядке по имени игрока.

```python
badmenton[badmenton.NOC == top_badmentons.index[0]].sort_values(by=['Athlete'])
```

# Какие три страны выйграли большинство медалей в последние годы (с 1984 по 2008)?

```python
top_three = oo[(oo['Edition'] <= 2008) & (oo['Edition'] >= 1984)].NOC.value_counts()
```

```python
print("ТОП 3 страны с большинством медалей:")
for number, item  in  enumerate(top_three[:3].items(), 1):
    print(f'{number}: Страна {item[0]}; Кол-во медалей: {item[1]}')
    
```

# Постройте график числа золотых медалей выйгранных США мужчинами иженщинами в атлетике?

# Постройте график 5 атлетов которые выйграли большинство золотых медалей

# HW #L10.1

```python
usa_athletes = oo[oo['NOC'] == 'USA']
#Группирую , отстековываю по кол-ву медалей, заменяю Nan на 0.0
tmp_table = usa_athletes.groupby(['Edition', 'Athlete', 'Discipline', 'Medal']).size().unstack('Medal')
tmp_table = tmp_table.fillna(value=0)
#Считаю кол-во медалей
tmp_table['Total'] = tmp_table.apply(lambda row: row.Bronze + row.Gold + row.Silver, axis = 1)

#Нахожу максимальное кол-во медалей в год и остовляю только 1 значение 
idx = tmp_table.groupby(['Edition'])['Total'].transform(max) == tmp_table['Total'] 
one_of_the_year = 0
for i, v in idx.items():
    if one_of_the_year != i[0] and v :
        one_of_the_year, idx[i] = i[0], True
    else:
        idx[i] = False
#Создаю новый датафрейм и обновляю индекс и его имя        
result = pd.DataFrame(tmp_table[idx])
result = result.reset_index()
result.keys().name = 'Index'
result
```
